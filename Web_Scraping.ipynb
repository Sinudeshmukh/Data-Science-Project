```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url  ='https://en.wikipedia.org/wiki/National_Stock_Exchange_of_India'
data = requests.get(url).text

soup = BeautifulSoup(data)

#get all the href links
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
## Taking the news text from space.com and print

page = requests.get("https://en.wikipedia.org/wiki/National_Stock_Exchange_of_India")
soup = BeautifulSoup(page.content , 'html.parser')

#find all tag in the soup.
heading = soup.find_all("p")

#...ptonic organization!
n=len(heading)
crypto = []
for x in range(n): 
    crypto.append(str.strip(heading[x].text))
```
